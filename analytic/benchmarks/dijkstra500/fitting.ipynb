```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
d = [
                            [
                                32,
                                0.6964
                            ],
                            [
                                64,
                                0.6967
                            ],
                            [
                                128,
                                0.7109
                            ],
                            [
                                256,
                                0.7161
                            ],
                            [
                                512,
                                0.7297
                            ],
                            [
                                1024,
                                0.7737
                            ],
                            [
                                2048,
                                0.953
                            ],
                            [
                                4096,
                                0.9919
                            ],
                            [
                                8192,
                                0.9943
                            ],
                            [
                                16384,
                                0.9985
                            ],
                            [
                                32768,
                                0.999
                            ]
                            ]
```

```python
def func5(x,a,b,c,d,e,f):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5

def func4(x,a,b,c,d,e):
    return a+b*x+c*x**2+d*x**3+e*x**4

def func3(x,a,b,c,d):
    return a+b*x+c*x**2+d*x**3

def func2(x,a,b,c):
    return a+b*x+c*x**2

def func(x, a, b):
    return a+b*x
```

```python
y = [m[1] for m in d]
x = np.linspace(5, 15, 11)
plt.plot(x[:], y[:], 'b-', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
f = func3
x1=x[:6]
y1=y[:6]
popt, pcov = curve_fit(f, x1, y1)
print(popt)
plt.plot(x1, y1, 'b-', label='data')
plt.plot(np.linspace(5, 10, 100), f(np.linspace(5, 10, 100), *popt), 'r-', label='regression')
# plt.plot(np.linspace(5, 7, 100), f(np.linspace(5, 7, 100), *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def ss(x):   
    return 0.34230952 + 0.16030635 *x -0.02420357 *x**2 +0.00124722*x**3
plt.plot(np.linspace(5, 10, 60), func3(np.linspace(5, 10, 60), *popt), 'b-', label='data')
plt.plot(np.linspace(5, 10, 60), ss(np.linspace(5, 10, 60)), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
