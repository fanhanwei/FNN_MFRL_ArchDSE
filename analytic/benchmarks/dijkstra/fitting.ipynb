```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
d = [
                            [
                                1,
                                0.1644
                            ],
                            [
                                2,
                                0.1846
                            ],
                            [
                                4,
                                0.7258
                            ],
                            [
                                8,
                                0.7259
                            ],
                            [
                                16,
                                0.7259
                            ],
                            [
                                32,
                                0.7267
                            ],
                            [
                                64,
                                0.7416
                            ],
                            [
                                128,
                                0.752
                            ],
                            [
                                256,
                                0.7773
                            ],
                            [
                                512,
                                0.8779
                            ],
                            [
                                1024,
                                0.9912
                            ],
                            [
                                2048,
                                0.9952
                            ],
                            [
                                4096,
                                0.9983
                            ]]
```

```python
def func1(x,a,b):
    return a+b*x

def func3(x,a,b,c,d):
    return a+b*x+c*x**2+d*x**3

def func4(x,a,b,c,d,e):
    return a+b*x+c*x**2+d*x**3+e*x**4

def func5(x,a,b,c,d,e,f):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5

def func6(x,a,b,c,d,e,f,g):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5+g*x**6

def func(x, **kwargs):
    y = 0
    for i, w in enumerate(kwargs.values()):
        y = y + float(w) * x**int(i)
    return y
```

```python
y = [m[1] for m in d]
x = np.linspace(0, len(y)-1, len(y))
```

```python
x1 = x[5:11]
y1 = y[5:11]
print(x1,y1)
```

```python

popt, pcov = curve_fit(func3, x1, y1)
print(popt)
plt.plot(x1, y1, 'b-', label='data')
plt.plot(np.linspace(5, 10, 60), func3(np.linspace(5, 10, 60), *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def ss(x):
    return 0.29682698 + 0.2267687*x + -0.0404381*x**2 +0.00247407*x**3
plt.plot(np.linspace(4, 10, 60), func3(np.linspace(4, 10, 60), *popt), 'b-', label='data')
plt.plot(np.linspace(4, 10, 60), ss(np.linspace(4, 10, 60)), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
