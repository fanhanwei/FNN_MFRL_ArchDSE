```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
d = [
                            [
                                0,
                                0.0934
                            ],
                            [
                                1,
                                0.4306
                            ],
                            [
                                2,
                                0.4308
                            ],
                            [
                                4,
                                0.4651
                            ],
                            [
                                8,
                                0.4993
                            ],
                            [
                                16,
                                0.5001
                            ],
                            [
                                32,
                                0.5002
                            ],
                            [
                                64,
                                0.5483
                            ],
                            [
                                128,
                                0.9281
                            ],
                            [
                                256,
                                0.9326
                            ],
                            [
                                512,
                                0.9941
                            ],
                            [
                                1024,
                                0.9969
                            ]]
```

```python
def func5(x,a,b,c,d,e,f):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5

def func4(x,a,b,c,d,e):
    return a+b*x+c*x**2+d*x**3+e*x**4

def func3(x,a,b,c,d):
    return a+b*x+c*x**2+d*x**3

def func2(x,a,b,c):
    return a+b*x+c*x**2

def func(x, a, b):
    return a+b*x
```

```python
y = [m[1] for m in d]
x = np.linspace(0, len(y)-1, len(y))
plt.plot(x[6:]-1, y[6:], 'b-', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# first interval
x1 = x[3:9]-1
y1 = y[3:9]
x_ = np.linspace(5, 7, len(y)*100)
print(x1, y1)
f = func5
popt, pcov = curve_fit(f, x1, y1)
print(popt)
plt.plot(x1[-3:], y1[-3:], 'b-', label='data')
plt.plot(x_, f(x_, *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# second interval
x2 = x[9:11]-1
y2 = y[9:11]
x_ = np.linspace(8, x2[-1]+1, len(y)*100)
print(x2, y2)
f = func
popt, pcov = curve_fit(f, x2, y2)
print(popt)
plt.plot(x2, y2, 'b-', label='data')
plt.plot(x_, f(x_, *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# 3rd interval
x2 = x[10:]-1
y2 = y[10:]
x_ = np.linspace(8, x2[-1]+1, len(y)*100)
print(x2, y2)
f = func
popt, pcov = curve_fit(f, x2, y2)
print(popt)
plt.plot(x2, y2, 'b-', label='data')
plt.plot(x_, f(x_, *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
