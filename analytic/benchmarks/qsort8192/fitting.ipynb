```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
d = [
                            [
                                0,
                                0.5263
                            ],
                            [
                                1,
                                0.9263
                            ],
                            [
                                2,
                                0.9529
                            ],
                            [
                                4,
                                0.9657
                            ],
                            [
                                8,
                                0.9721
                            ],
                            [
                                16,
                                0.9765
                            ],
                            [
                                32,
                                0.98
                            ],
                            [
                                64,
                                0.9832
                            ],
                            [
                                128,
                                0.986
                            ],
                            [
                                256,
                                0.9889
                            ],
                            [
                                512,
                                0.9914
                            ],
                            [
                                1024,
                                0.9949
                            ]]
```

```python
def func5(x,a,b,c,d,e,f):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5

def func4(x,a,b,c,d,e):
    return a+b*x+c*x**2+d*x**3+e*x**4

def func3(x,a,b,c,d):
    return a+b*x+c*x**2+d*x**3

def func2(x,a,b,c):
    return a+b*x+c*x**2

def func(x, a, b):
    return a+b*x
```

```python
y = [m[1] for m in d]
x = np.linspace(0, len(y)-1, len(y))
plt.plot(x[6:]-1, y[6:], 'b-', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x1 = x[6:]-1
y1 = y[6:]
x_ = np.linspace(5, x1[-1], len(y)*100)
print(x1, y1)
f = func3
popt, pcov = curve_fit(f, x1, y1)
print(popt)
plt.plot(x1, y1, 'b-', label='data')
plt.plot(x_, f(x_, *popt), 'r-', label='regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
